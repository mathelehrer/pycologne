```python
import z3.z3 as z3
```

Wir wollen das folgende Krypto-Arithemtikum lÃ¶sen:
~~~
    DACS
  - INFO
 -------
    NOOO
~~~

```python
ziffern = z3.Ints("D A C S I N F O")
D, A, C, S, I, N, F, O = ziffern

sol = z3.Solver()
for i, ziffer in enumerate(ziffern):
    sol.add(ziffer >= 0)
    sol.add(ziffer < 10)
    for andere in ziffern[i+1:]:
        sol.add(ziffer != andere)
```

```python
DACS = 1000 * D + 100 * A + 10 * C + S
INFO = 1000 * I + 100 * N + 10 * F + O
NOOO = 1000 * N + 100 * O + 10 * O + O
sol.add(DACS - INFO == NOOO)
```

```python
sol.check()
```

```python
mod = sol.model()   
mod.eval(DACS) - mod.eval(INFO), mod.eval(NOOO)
```

```python
while sol.check() == z3.sat:
    mod = sol.model()
    print(mod.eval(DACS) - mod.eval(INFO), mod.eval(NOOO))
    sol.add(DACS != mod.eval(DACS))
```
